```python
import polars as pl
import os

def summarize_parquet_data_polars(file_path):
    """
    Polarsを使用してParquetファイルを読み込み、データの概要を表示する。
    """
    # 1. データの読み込み
    # PolarsはデフォルトでParquetを高速に読み込めます
    df = pl.read_parquet(file_path)

    print("--- Basic Information ---")
    # shapeはPandasと同様 (rows, cols)
    print(f"Shape: {df.shape}")

    print("\n--- Data Types (Schema) ---")
    # info()の代わりにschemaを確認
    print(df.schema)

    print("\n--- Missing Values Count ---")
    # 列ごとの欠損値を集計
    print(df.null_count())

    print("\n--- Descriptive Statistics ---")
    # describe()で統計量を算出
    print(df.describe())

    return df

path_to_file = '../data/Esprit/RXYP615FC'
parquet_name = 'full_result.parquet'

# 実行
df = summarize_parquet_data_polars(os.path.join(path_to_file, parquet_name))

# データの先頭を表示
print("\n--- Head ---")
print(df.head())
```

## EV 開度比の関数

### 2026-３-31以降の方法
2026-3-17の方法にCvの補正（1.566）で割って開度比を算出する

```python
import numpy as np


def evalve_date_open_ratio(
    m: float | np.ndarray,
    rho: float | np.ndarray,
    delta_p: float | np.ndarray,
    c_v: float | np.ndarray,
) -> float | np.ndarray:
    """電子膨張弁の開度比を算出する
    Parameters
    ----------
    m : float or array_like
        質量流量 [kg/s]
    rho : float or array_like
        密度 [kg/m³]
    delta_p : float or array_like
        差圧 [Pa]
    c_v : float or array_like
        Cv値
    Returns
    -------
    float or ndarray
        開度比 r
    """
    r = m / c_v * np.sqrt(1730.882 / (rho * delta_p)) * 10**3
    return r
```

## Ph線図のプロット

```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import CoolProp.CoolProp as CP
import yaml

import pandas as pd


# 1. ヘルパー関数: エンタルピー計算 (get_h)
def get_h(T_C, P_MPa, fluid):
    if pd.isna(T_C) or pd.isna(P_MPa):
        return None
    T_K = T_C + 273.15
    P_Pa = P_MPa * 1e6
    try:
        return CP.PropsSI('H', 'T', T_K, 'P', P_Pa, fluid) / 1000.0
    except ValueError:
        return None
    
# 飽和線と等温線
def add_ph_background(fig, fluid):
    Tc = CP.PropsSI('Tcrit', fluid)
    T_min = 223.15  # -50 degC
    T_range = np.linspace(T_min, Tc - 0.1, 100)

    h_l, p_l, h_v, p_v = [], [], [], []
    for T in T_range:
        h_l.append(CP.PropsSI('H', 'T', T, 'Q', 0, fluid) / 1000.0)
        p_l.append(CP.PropsSI('P', 'T', T, 'Q', 0, fluid) / 1e6)
        h_v.append(CP.PropsSI('H', 'T', T, 'Q', 1, fluid) / 1000.0)
        p_v.append(CP.PropsSI('P', 'T', T, 'Q', 1, fluid) / 1e6)
    
    fig.add_trace(go.Scatter(x=h_l, y=p_l, mode='lines', line=dict(color='black', width=2), showlegend=False, hoverinfo='skip'))
    fig.add_trace(go.Scatter(x=h_v, y=p_v, mode='lines', line=dict(color='black', width=2), showlegend=False, hoverinfo='skip'))

    T_celsius = [0, 20, 40, 60, 80, 100]
    p_range = np.geomspace(0.5e6, 5.0e6, 50)

    for Tc_val in T_celsius:
        T_kelvin = Tc_val + 273.15
        h_iso, p_iso = [], []
        for p in p_range:
            try:
                h = CP.PropsSI('H', 'T', T_kelvin, 'P', p, fluid) / 1000.0
                h_iso.append(h)
                p_iso.append(p / 1e6)
            except: continue
        
        fig.add_trace(go.Scatter(x=h_iso, y=p_iso, mode='lines', name=f'{Tc_val}°C',
                                 line=dict(color='green', width=1, dash='dot'), opacity=0.4, showlegend=False, hoverinfo='skip'))
        if h_iso:
            fig.add_annotation(x=h_iso[-1], y=p_iso[-1], text=f"{Tc_val}°C", showarrow=False, font=dict(color="green", size=10), xanchor="left")


# メインの回路描画関数 (Streamlit関数を排除)
def create_ph_diagram(current_row, config):
    fluid = config.get('fluid', 'R410A')
    fig_ph = go.Figure()
    add_ph_background(fig_ph, fluid)

    # 1. PolarsからPandasへ変換 (計算直前)
    current_row_pd = current_row.to_pandas()

    points_coords = {}

    # 点のプロット
    for node_id, info in config['nodes'].items():
        t_val = current_row_pd.get(info.get('temp'))
        p_val = current_row_pd.get(info.get('press_ref'))

        print(t_val,p_val)
        
        if t_val is not None and p_val is not None:
            h = get_h(t_val, p_val, fluid) 
            if h is not None:
                points_coords[node_id] = (h, p_val)
                fig_ph.add_trace(go.Scatter(x=[h], y=[p_val], mode='markers+text', 
                                            name=info['label'], text=[info['label']], textposition="top center",
                                            marker=dict(size=10, symbol=info.get('symbol', 'circle')),
                                            customdata=[[t_val, p_val]],
                                            hovertemplate="<b>%{text}</b><br>H: %{x:.1f} kJ/kg<br>P: %{y:.2f} MPa<br>T: %{customdata[0]:.1f} °C<extra></extra>"))

    # 線のプロット
    for conn in config['connections']:
        if conn['start'] in points_coords and conn['end'] in points_coords:
            h_s, p_s = points_coords[conn['start']]
            h_e, p_e = points_coords[conn['end']]
            fig_ph.add_trace(go.Scatter(x=[h_s, h_e], y=[p_s, p_e], mode='lines', name=conn.get('label'),
                                        line=dict(color=conn.get('color'), width=conn.get('width'), dash=conn.get('dash')), opacity=0.8))

    fig_ph.update_layout(
        title=f"Ph Diagram - {config.get('model_name')}",
        xaxis_title="Enthalpy [kJ/kg]", yaxis_title="Pressure [MPa]", yaxis_type="log",
        xaxis=dict(range=[100, 700]), yaxis=dict(range=[np.log10(0.3), np.log10(6.0)]),
        height=800, template="plotly_white", showlegend=True
    )
    
    return fig_ph
```

```python
circuit_yaml = 'RXYP615FC_circuit.yml'
path_to_file = '../data/Esprit/RXYP615FC/'
circuit_yaml_path = os.path.join(path_to_file,circuit_yaml)

import yaml
with open(circuit_yaml_path, 'r', encoding='utf-8') as f:
    circuit_dict = yaml.safe_load(f)
```

```python
current_row = df[10]

# グラフ生成    
fig = create_ph_diagram(current_row=current_row, config=circuit_dict)

# --- 出力方法の選択 ---
# 方法A: ブラウザで即座に表示
fig.show()
```
