```python
import numpy as np
import matplotlib.pyplot as plt
import CoolProp.CoolProp as CP

def plot_ph_diagram(fluid_name, color):
    # 臨界点の取得
    T_crit = CP.PropsSI('Tcrit', fluid_name)
    P_crit = CP.PropsSI('Pcrit', fluid_name)
    
    # 飽和温度の範囲（トリプルポイント付近から臨界点手前まで）
    T_min = CP.PropsSI('Tmin', fluid_name)
    T_range = np.linspace(T_min + 0.1, T_crit - 0.1, 100)
    
    # 飽和液線 (Quality = 0)
    h_l = [CP.PropsSI('H', 'T', T, 'Q', 0, fluid_name) / 1000 for T in T_range] # kJ/kg
    p_l = [CP.PropsSI('P', 'T', T, 'Q', 0, fluid_name) / 1e6 for T in T_range]  # MPa
    
    # 飽和蒸気線 (Quality = 1)
    h_v = [CP.PropsSI('H', 'T', T, 'Q', 1, fluid_name) / 1000 for T in T_range] # kJ/kg
    p_v = [CP.PropsSI('P', 'T', T, 'Q', 1, fluid_name) / 1e6 for T in T_range]  # MPa
    
    # 描画
    plt.plot(h_l, p_l, color=color, lw=2, label=f'{fluid_name} Sat. Liquid')
    plt.plot(h_v, p_v, color=color, lw=2, ls='--', label=f'{fluid_name} Sat. Vapor')

# グラフ設定
plt.figure(figsize=(10, 7))

# R32 と R410A をプロット
plot_ph_diagram('R32', 'blue')
plot_ph_diagram('R410A', 'red')

# 仕上げ
plt.yscale('log') # 圧力は対数軸が一般的
plt.xlabel('Specific Enthalpy [kJ/kg]')
plt.ylabel('Pressure [MPa]')
plt.title('P-h Diagram: R32 vs R410A')
plt.grid(True, which="both", ls="-", alpha=0.5)
plt.legend()

# 軸の範囲調整（見やすくするため）
plt.xlim(100, 600)
plt.ylim(0.8, 10)

plt.show()
```

```python
import CoolProp.CoolProp as CP

fluid = "R410A"
temp_c = 53.0
temp_k = temp_c + 273.15  # ℃からKへ変換

# 1. 飽和液圧力 (Quality = 0): Bubble point
p_bubble = CP.PropsSI('P', 'T', temp_k, 'Q', 0, fluid)

# 2. 飽和蒸気圧力 (Quality = 1): Dew point
p_dew = CP.PropsSI('P', 'T', temp_k, 'Q', 1, fluid)

print(f"--- {fluid} (温度: {temp_c} ℃) の飽和圧力 ---")
print(f"飽和液圧 (Bubble point): {p_bubble / 1e6:.4f} MPa")
print(f"飽和蒸気圧 (Dew point): {p_dew / 1e6:.4f} MPa")
print(f"圧力差 (Gliding): {(p_bubble - p_dew):.1f} Pa")
```

```python
# 1. 飽和液の比エンタルピー (Quality = 0)
# p_bubble の状態に相当します
h_l = CP.PropsSI('H', 'T', temp_k, 'Q', 0, fluid)

# 2. 飽和蒸気の比エンタルピー (Quality = 1)
# p_dew の状態に相当します
h_v = CP.PropsSI('H', 'T', temp_k, 'Q', 1, fluid)

# 蒸発潜熱 (Latent Heat) の計算
h_fg = h_v - h_l

print(f"--- {fluid} (温度: {temp_c} ℃) の比エンタルピー ---")
print(f"飽和液比エンタルピー (h_l): {h_l / 1000:.2f} kJ/kg")
print(f"飽和蒸気比エンタルピー (h_v): {h_v / 1000:.2f} kJ/kg")
print(f"蒸発潜熱 (h_fg): {h_fg / 1000:.2f} kJ/kg")
```
